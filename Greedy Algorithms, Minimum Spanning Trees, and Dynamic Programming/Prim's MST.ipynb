```python
import numpy as np
first=True
graph={}
with open('input5.txt') as f:
    line=f.readline()
    line=line.replace('\n',"")
    line=line.split(" ")
    n=int(line[0])
    m=int(line[1])
    lines=f.readlines()
    for i in lines:
        E=i.strip().split(" ")
        u,v,w=int(E[0]),int(E[1]),int(E[2])
        if u in graph.keys():
            graph[u][v]=w
        else:
            graph[u]={}
            graph[u][v]=w
        if v in graph.keys():
            graph[v][u]=w
        else:
            graph[v]={}
            graph[v][u]=w
```

```python
# book keeping
seen,cost,initial_vertex,tree={},0,1,[]
# initializing the d vector containing the distance to the seen tree
d=np.array([np.inf]*n)
# initializing the pi vector containing the neighbours from the MST
pi=np.array([0]*n)
```

```python
# adding initial_vertex to the seen repo and making changes to the d vector
seen[initial_vertex]=True
for i in graph[initial_vertex].keys():
    if d[i-1]>graph[initial_vertex][i]:
        d[i-1]=graph[initial_vertex][i]
        pi[i-1]=initial_vertex
```

```python
# lets get the iterator started, such that we include one node at a time into the seen dictionary
while len(seen.keys())<n:
    child=np.argmin(d)+1
    parent=pi[child-1]
    seen[child]=True
    tree=tree+[[parent,child]]
    for i in graph[child].keys():
        if i not in seen.keys():
            if d[i-1]>graph[child][i]:
                d[i-1]=graph[child][i]
                pi[i-1]=child
    d[child-1]=np.inf
    cost+=graph[parent][child]
    
```

```python
print(cost)
for i in tree:
  if i[0]<i[1]:
    print(i[0],i[1])
  else:
    print(i[1],i[0])
```
