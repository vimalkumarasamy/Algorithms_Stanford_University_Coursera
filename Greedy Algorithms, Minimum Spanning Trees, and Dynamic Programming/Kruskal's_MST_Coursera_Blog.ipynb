```
filename ='gdrive/My Drive/Algorithms/clustering_big.txt'
```

```
firstline=True
nodes_to_leader={}
leader_to_mem={}
for f in open(filename):
  if firstline:
    firstline=False
    pass
  else:
    f=f.replace("\n","")
    f=f.replace(" ","")
    nodes_to_leader[int(f,2)]=int(f,2)
    leader_to_mem[int(f,2)]=[int(f,2)]
```

```
def friends(node,nodes):
  from itertools import combinations 
  frens=[]
  mask=[1<<i for i in range(24)]
  comb = combinations(mask, 2)
  # print(mask)
  for i in mask:
    if node^i in nodes:
      frens+=[node^i]
  for i in comb:
    if (node^(1<<i[0]))^(1<<i[1]) in nodes:
      frens+=[(node^(1<<i[0]))^(1<<i[1])]
  return(frens)
```

```
m=0
for node in nodes_to_leader:
  for friend in friends(node,nodes_to_leader):
    if nodes_to_leader[node]!=nodes_to_leader[friend]:
      for i in leader_to_mem[nodes_to_leader[friend]]:
        nodes_to_leader[i]=nodes_to_leader[node]
      # leader_to_mem[nodes_to_leader[node]]=leader_to_mem[nodes_to_leader[friend]]+leader_to_mem[nodes_to_leader[node]]
      # del(leader_to_mem[nodes_to_leader[friend]])
  if m%100==0:
    print(m,' nodes processed')
  m=m+1
```
